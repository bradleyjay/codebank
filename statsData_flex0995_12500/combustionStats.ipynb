```python
### determine convergence ###

'''

Rolling mean is used here to calculate a windowed mean. As the rolling mean stops 
changing, the data's average value is becoming constant. This means the data
isn't changing.

Likewise, as the rolling mean becomes more constant, standard deviation drops.
Standard deviation is the average distance of values in the window from that 
window's mean. So, less variation when N is large enough to catch a changing
mean; the mean becomes more representative, relatively.

Convergence is reached once the (1) mean stops changing and (2) the standard dev
drops to a 'sufficent' level.

'''
N = 100  # sample window size
df['rm_Broil'] = df['Broil'].rolling(N).mean() #get mean of window N
df['std_broil']= df['Broil'].rolling(N).std()  #get standard dev of that window
df[['Iteration','Broil']].plot(x='Iteration', title = 'Broil TC Raw',xlim=(0,3000)) # display raw Broil TC
df[['Iteration','rm_Broil']].plot(x='Iteration', title='Broil Rolling Mean, N = 100',xlim=(0,3000),ylim=(550,700))
df[['Iteration','std_broil']].plot(x='Iteration', title='Broil TC STD, N = 100',xlim=(0,3000),ylim=(3,25))

# Explore this for various values of N, plot all series on one graph to find what
# TC we waited for!
```
